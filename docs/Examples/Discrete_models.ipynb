## SEIpRpS model
This is a Susceptible-Exposed-Infectious-Recovered-Persistent-Susceptible model. It is a discrete time model that can be used to simulate the spread of infectious diseases in a population. The model is based on the assumption that individuals can move between different compartments (Susceptible, Exposed, Infectious, Recovered, Persistent) over time. The model uses a set of differential equations to describe the dynamics of the disease spread and the interactions between the compartments.

````python
modelseiprps = DM.SEIpRpS()
````

````python
tsteps = [0,50]
modelseiprps([1000, 1, 0, 0], tsteps, 1002, {'beta': 2, 'r': 1, 'e': 1, 'b': 0, 'delta': 0.5})
modelseiprps.plot_traces()
````
