# Simulating Discrete models

```python
from epimodels.discrete import models as DM
from IPython.display import Markdown as md
```

## SIS model

```python
modelsis = DM.SIS()
md(str(modelsis))
```

```python
modelsis([0, 1, 1000], [0,50], 1001, {'beta': 2, 'gamma': 1})
modelsis.plot_traces()
```

## SIR Model

```python
sirmodel = DM.SIR()
```

```python
md(str(sirmodel))
```

```python
sirmodel([1000, 1, 0], [0,500], 1001, {'beta': .2, 'gamma': .1})
sirmodel.plot_traces()
```

## SEIpRpS model
This model assumes that a fraction $\delta$ of infectious individuals acquire full immunity while the remaining $(1-\delta)$ returns to the susceptible stage. The model is:

```python
modelseiprps = DM.SEIpRpS()
```

```python
md(str(modelseiprps))
```

```python
tsteps = [0,50]
modelseiprps([1000, 0, 1, 0], tsteps, 1001, {'beta': 1, 'r': .3, 'e': 1, 'b': 0, 'delta': 0.6})
modelseiprps.plot_traces()
```
